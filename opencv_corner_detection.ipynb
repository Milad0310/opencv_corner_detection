```python
#Import Library
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Harris Corner Detection

```python
#Read Image
chess_img = cv2.imread("flat_chessboard.png")

#convert Image Into Float & Gray
gray = cv2.cvtColor(chess_img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

#Detect Corner with Harris Corner Detection
dst = cv2.cornerHarris(gray, 2, 3, 0.04)

#Plot Detected Corner in Img
chess_img[dst>0.01*dst.max()]=[0,0,0]
plt.imshow(chess_img)
```

# Shi-Tomasi Corner Detector

```python
#Read Image
chess_img = cv2.imread("flat_chessboard.png")

#convert Image Into Float & Gray
gray = cv2.cvtColor(chess_img, cv2.COLOR_BGR2GRAY)

#Detect Corner with Shi-Tomasi Corner Detector
corners = cv2.goodFeaturesToTrack(gray, 0,0.01,10)

corners = np.int0(corners)

for i in corners:
    x, y = i.ravel() 
    cv2.circle(chess_img,(x,y),3,0,-1)
    
plt.imshow(chess_img)
```
